```python
import csv
def file_read():
    with open('D:/교통/DM_RUTBY_PASGR_T.dat') as file:
        while True:
            line = file.readline()
            line_list=line.split('|')
            if line == '':
                break
            yield line_list                                   
f=open('DM_RUTBY_PASGR_T_1.csv','w',newline='')
wr=csv.writer(f)
for i in file_read():
    if(i[11]=='27'):
        
        wr.writerow(i)
f.close()
```
